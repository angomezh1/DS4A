## Estructuras de control

### Ciclos FOR

Los ciclos For en Python son simples. La estructura basica esta dada por **`for <var-name> in <iterable> :`**
**Importante** finalizando el for siempre debe de ir **:**, es obligatorio y para los constructores del flujo de control son obligaci贸n. 

El bloque que sera iterado esta determinado por la tabulaci贸n de las lineas siguientes del for.   
   
La linea en blanco despues de las dos sentencias no es obligatoria, pero es considerada buena practica. Las lineas luego del ciclo for no llevan tabulaci贸n, por lo tanto ya estan fuera del ciclo.

`````
string_list = [ "Juan", "is", "1.70", "metres", "high"]
`````

`````
for elem in string_list : 
    print( elem )
    print( len(elem), 'characters\n' )
    
print( "The loop is done")
`````

`````
for idx in range(len(string_list)):
  print(idx)
  print(string_list[idx])
  print(len(string_list[idx]))
`````
